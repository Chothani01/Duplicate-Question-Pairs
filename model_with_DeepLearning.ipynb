```python
df['question1'] = df['question1'].apply(preprocess)
df['question2'] = df['question2'].apply(preprocess)
```

```python
questions = list(df['question1']) + list(df['question2'])
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(questions)
```

```python
df['question1'] = tokenizer.texts_to_sequences(df['question1'])
df['question2'] = tokenizer.texts_to_sequences(df['question2'])
```

```python
x = df[['question1', 'question2']]
y = df['is_duplicate']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
x_train['question1'].info()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_train_input1 = pad_sequences(x_train['question1'], padding='post', maxlen=12)
x_train_input2 = pad_sequences(x_train['question2'], padding='post', maxlen=12)
x_test_input1 = pad_sequences(x_test['question1'], padding='post', maxlen=12)
x_test_input2 = pad_sequences(x_test['question2'], padding='post', maxlen=12)
```

```python
print(len(tokenizer.word_index))
```

```python
from tensorflow.keras.layers import Concatenate, Dense, Dropout, LSTM, Input, Bidirectional, Embedding
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
```

```python
input1 = Input(shape=(12,))
x1 = Embedding(input_dim=95603, output_dim=100)(input1)

input2 = Input(shape=(12,))
x2 = Embedding(input_dim=95603, output_dim=100)(input2)

combined = Concatenate()([x1, x2])
x3 = Bidirectional(LSTM(100, kernel_regularizer=l2(0.001)))(combined)
x4 = Dropout(0.5)(x3)

output = Dense(1, activation='sigmoid')(x4)

model = Model(inputs=[input1, input2], outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit([x_train_input1, x_train_input2], y_train, epochs=10, batch_size=32, validation_data=([x_test_input1, x_test_input2], y_test))
# We convert inputs and oututs in list because model not support object datatype
```
