```python
new_df = df.dropna(axis=0)
new_df
```

```python
x = new_df[['question1', 'question2']]
y = df['is_duplicate']
x, y
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2000)
questions = list(new_df['question1']) + list(new_df['question2'])
# [1,2,3]+[4,5,6] = [1,2,3,4,5,6]

q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(), 2)
```

```python
temp1_df = pd.DataFrame(q1_arr, index=new_df.index)
temp2_df = pd.DataFrame(q2_arr, index=new_df.index)
temp_df = pd.concat([temp1_df, temp2_df], axis=1)
temp_df.shape
```

```python
temp_df
```

```python
temp_df['is_duplicate'] = new_df['is_duplicate']
temp_df
```

```python
x = temp_df.drop(columns=['is_duplicate'])
y = temp_df['is_duplicate']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
df = RandomForestClassifier()
df.fit(x_train, y_train)
y_pred = df.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
xgb = XGBClassifier()
xgb.fit(x_train.values, y_train.values) # because
y_pred = xgb.predict(x_test.values)
accuracy_score(y_test, y_pred)
```
